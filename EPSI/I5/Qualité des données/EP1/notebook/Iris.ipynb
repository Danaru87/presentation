## Visualiser nos data : 

```python
# Lib de visualition
import seaborn as sns
import matplotlib.pyplot as plt
# Pour que les graphique s'affichent dans le notebook
%matplotlib inline 
```

```python
iris.plot() # Va afficher les données du DataFrame (avec Pandas)
```

- L'abscisse est juste l'index du DataFrame --> *"x"*
- L'ordonnées est la mesure des variable affiché --> *"y"*

```python
iris[['SepalLengthCm', 'SepalWidthCm']].plot()
```

```python
# Visualiser un nuage de point (Scatter)
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
iris.head()
```

```python
# Utilisation de Seaborn (lib de visualittion via sns.)
sns.jointplot(x="PetalLengthCm", y="SepalWidthCm", data=iris)
```

```python
# les boxplot (aka boite à moustache) 
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
sns.pairplot(iris, hue="Species")
```
